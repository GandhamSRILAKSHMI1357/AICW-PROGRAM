```python
Imdb_df['Rate'] = pd.to_numeric(Imdb_df['Rate'], errors = 'coerce')
Imdb_df.info()
```

```python
for i in range(0,Imdb_df.shape[0]):
    Imdb_df['Votes']= Imdb_df['Votes'].astype(str).str.replace(',','')
    
Imdb_df['Votes']
```

```python
Imdb_df['Votes'] = pd.to_numeric(Imdb_df['Votes'], errors = 'coerce')
Imdb_df.info()
```

```python
KingKong_data = Imdb_df.loc[Imdb_df.Name=='KingKong']
KingKong_data
```

```python
Imdb_df = Imdb_df.drop_duplicates()
print(Imdb_df.shape)
```

```python
Imdb_df.isnull().sum()
```

```python
Imdb_df['Rate']=Imdb_df['Rate'].replace(np.nan,0)
Imdb_df['Votes']=Imdb_df['Votes'].replace(np.nan,0)
```

```python
Imdb_df['Certificate'] = Imdb_df['Certificate'].replace(np.nan, "None")
Imdb_df['Nudity'] = Imdb_df['Nudity'].replace(np.nan, "None")
Imdb_df['Violence'] = Imdb_df['Violence'].replace(np.nan, "None")
Imdb_df['Profanity'] = Imdb_df['Profanity'].replace(np.nan, "None")
Imdb_df['Alcohol'] = Imdb_df['Alcohol'].replace(np.nan, "None")
Imdb_df['Frightening'] = Imdb_df['Frightening'].replace(np.nan, "None")
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,5))
sns.set(style='dark')
sns.countplot(x="Date", data = Imdb_df)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.set(style='dark')
sns.histplot(data = Imdb_df, x='Rate', color='indigo', binwidth=1)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x=Imdb_df.Rate,y =Imdb_df.Votes);
plt.show()
```

```python
Top_Rated = Imdb_df.nlargest(10,'Rate')
Top_Rated
```

```python
plt.figure(figsize=(5,6))
plot = sns.countplot(x='Type',data=Top_Rated)
plt.show()
```
